```python
def variance(data, ddof=0): #ddof = 0 for population, =1 for sample variance 
    n = len(data)
    if(n != 0):
        mean = sum(data) / n
        return sum((x - mean) ** 2 for x in data) / (n - ddof)
    else:
        return 144
```

```python
def st_dev(data):
    var = variance(data)
    std_dev = math.sqrt(var)
    return std_dev
```

```python
def sd_assigner(df, st_dev):
    
    df["st_dev"] = st_dev
    for x in df.index:
        if(abs(df["%change"][x]) > st_dev):
            df[">st_dev"][x] = 1
    
    return df
```

```python
def st_dev_calculator(price_df):
    temp_df = price_df
    slice_df = pd.DataFrame()
    result_df = pd.DataFrame()
    temp_df["price_dif"] = 0.0
    temp_df["%change"] = 0.0
    temp_df["st_dev"] = 0.0
    temp_df[">st_dev"] = 0
    
    obs = []
    index_stamp = 1
    for x in temp_df.index:
        if(x != 0):
            x_price = temp_df["price"][x]
            pre_price = temp_df["price"][x-1]
            
            if(temp_df["ID"][x] == temp_df["ID"][x-1]):
                if(temp_df["date_time"][x].month == temp_df["date_time"][x-1].month): #change according to month checking
                    temp_df["price_dif"][x] = x_price-pre_price
                    x_price_dif = temp_df["price_dif"][x]
                    temp_df["%change"][x] = x_price_dif/pre_price
                    x_rate = temp_df["%change"][x]
                    obs.append(abs(x_rate))
                    
                else:
                    #calculating for new month's #1
                    temp_df["price_dif"][x] = x_price-pre_price
                    x_price_dif = temp_df["price_dif"][x]
                    temp_df["%change"][x] = x_price_dif/pre_price
                    x_rate = temp_df["%change"][x]
                    
                    #st_dev operation for previous month
                    sd = st_dev(obs)
                    
                    interval = list(range(index_stamp, x))
                    slice_df = temp_df[temp_df.index.isin(interval)]
                    
                    slice_df_v2 = sd_assigner(slice_df, sd)
                    result_df = result_df.append(slice_df_v2)
                    
                    obs.clear()
                    index_stamp = x
                    
                    #cont'd for this month
                    obs.append(abs(x_rate))
            else:
                sd = st_dev(obs)
                
                interval = list(range(index_stamp, x))
                slice_df = temp_df[temp_df.index.isin(interval)]
                    
                slice_df_v2 = sd_assigner(slice_df, sd)
                result_df = result_df.append(slice_df_v2)
                
                obs.clear()
                index_stamp = x+1
    
    return result_df
```

```python
price_df_v2 = st_dev_calculator(price_df)
```
